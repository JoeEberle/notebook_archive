```python
# Import all of the necessary Libraries 
import logging
import configparser

# Initialize the confuration settings parser engine to get and set parameters 
config = configparser.ConfigParser()
config_filename = 'C:/Users/JAE670/config.ini'
cfg = config.read(config_filename)   

path_to_log_directory = config.get('event_logging','configuration_file_directory')
log_filename = path_to_log_directory + config.get('event_logging','log_file_name')

# Create and configure the Logger 
log_format = "%(levelname)s %(asctime)s %(created)f %(filename)s %(levelno)s %(lineno)d %(message)s %(msecs)d "
logging.basicConfig(filename = log_filename,
                   level = logging.DEBUG,
                   format = log_format,
                   filemode = 'w')  
logger = logging.getLogger() 

log_level = logger.level
logger.debug('Logger Initialize to level:{}'.format(log_level))
logger.info('Logger Initialized successfully to logging level:{}'.format(log_level))
logger.info('Logger Initialized write to log file:{}'.format(log_filename))
```
