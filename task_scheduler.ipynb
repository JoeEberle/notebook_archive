```python
import schedule
import time 
```

# Example Jobs and schedule

```python
def job():
    print('Scheduled Job')
    
def task():
    print('Scheduled task')   
    
def execute_ETL():
    print('Scheduled execute ETL')       
```

```python
schedule.every(2).seconds.do(job)
schedule.every(5).seconds.do(task)
schedule.every().day.at('03:00').do(execute_ETL)

while True:
    schedule.run_pending()
    time.sleep(1)
```
