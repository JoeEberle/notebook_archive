```python
import pandas as pd
```

```python
dialogue_df = pd.read_excel('C:\infrastructure\excel\dialogue.xlsx')
dialogue_df.columns = ["dialog"]
dialogue_df.head(10)
```

```python
def classify_dialogue_type(dialog):
    dialog_text = str(dialog)
    type = 'comment'
    # Start with the names the GPPC explicitly says not to LOAD
    if dialog_text.find('?') > -1:
        type = 'question'  
    elif dialog_text.find('|') > -1:
        type = 'response' 
    return  type                                  
    
```

```python
# Persist the dataset 
dialogue_df['dialogue_type'] = dialogue_df.apply(lambda x: classify_dialogue_type(x['dialog']), axis=1)
dialogue_df.head(30)
```

```python
dialogue_df.to_excel('C:\infrastructure\excel\dialog.xlsx')
dialogue_df.to_json('C:\infrastructure\excel\dialog.json')
```
