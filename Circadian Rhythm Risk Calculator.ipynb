```python
def interpret_stop_breathing_risk(row):
    if (row['Has anyone ever observed that you stop breathing during your sleep?'] == 1.0):  
        row['Sleep Stop Breathing Risk']= 1
    else: row['Sleep Stop Breathing Risk'] = 0      
    return row

def interpret_trouble_falling_asleep(row):
    if (row['Over the last 2 weeks, how many days have you had trouble falling asleep or staying asleep or sleeping too much?'] == 1.0):  
        row['Trouble Falling Asleep']= 1
    else: row['Trouble Falling Asleep'] = 0      
    return row

def interpret_low_energy(row):
    if (row['Over the last 2 weeks, how many days have you felt tired or had little energy?'] == 1.0):  
        row['Low Energy']= 1
    else: row['Low Energy'] = 0      
    return row

def interpret_sleeping_during_day(row):
    #  Interpret sleeping during day time hours
    if (row['During the past 30 days how many days did you unintentionally falling asleep during the day?'] == 1.0):  
        row['Daytime Sleeping']= 1
    else: row['Daytime Sleeping'] = 0      
    return row

def interpret_sleep_driving(row):
    if (row['During the past 30 days, have you ever nodded off or fallen asleep, even just for a brief moment, while driving?'] == 1.0 ):  
        row['Sleep Driving']= 1
    else: row['Sleep Driving'] = 0      
    return row

def interpret_snore_loudly(row):
    if (row['Do you snore loudly and wake people up?'] == 1.0):  
        row['Snore Loudly']= 1
    else: row['Snore Loudly'] = 0      
    return row
          
def interpret_fasting(row):
    if (row['Do you periodically fast 24 hours or longer or fast to lose weight?'] == 1.0):  
        row['Fasting']= 1
    else: row['Fasting'] = 0      
    return row
              
    
              
def sum_CIRC_risk_Factors(row):
    row['Circadian Rythm Risk Factors'] =  row['Sleep Stop Breathing Risk'] + row['Trouble Falling Asleep'] + \
    row['Low Energy'] + row['Daytime Sleeping'] + \
    row['Sleep Driving'] + row['Snore Loudly']  + row['Fasting']
    return row  
              
def sum_CIRC_risk_Score(row):
    row['Circadian Rythm Score'] =  (row['Sleep Stop Breathing Risk'] * 1.0) + (row['Trouble Falling Asleep'] * 1.0)+ \
    (row['Low Energy'] * 1.0) + (row['Daytime Sleeping'] * 1.0) + \
    (row['Sleep Driving'] * 1.0) + (row['Snore Loudly'] * 1.0) + (row['Fasting'] * 1.0)
    return row  

def sum_CIRC_risk_Score_wt(row):
    row['Circadian Rythm Score Weighted'] =  (row['Sleep Stop Breathing Risk'] * 0.098) + (row['Trouble Falling Asleep'] * 0.2)+ \
    (row['Low Energy'] * 0.24) + (row['Daytime Sleeping'] * 0.21) + \
    (row['Sleep Driving'] * 0.03) + (row['Snore Loudly'] * -0.014) + (row['Fasting'] * 0.3)
    return row  


              
def sum_CIRC_risk_Strat(row):
    row['Circadian Rythm Risk'] =  row['Circadian Rythm Risk Factors'] / 7
    return row

def sum_CIRC_risk_Cohort(row):
    perc_cohort = '' 
    if (row['Sleep Stop Breathing Risk'] == 1.0):
        perc_cohort = perc_cohort + 'Sleep Stop Breathing Risk, '
    if (row['Trouble Falling Asleep'] == 1.0):
        perc_cohort = perc_cohort + 'Trouble Falling Asleep, '
    if (row['Low Energy'] == 1.0):
        perc_cohort = perc_cohort + 'Low Energy, '       
    if (row['Daytime Sleeping'] == 1.0):
        perc_cohort = perc_cohort + 'Daytime Sleeping, ' 
    if (row['Sleep Driving'] == 1.0):
        perc_cohort = perc_cohort + 'Sleep Driving, '  
    if (row['Snore Loudly'] == 1.0):
        perc_cohort = perc_cohort + 'Snore Loudly, ' 
    if (row['Fasting'] == 1.0):
        perc_cohort = perc_cohort + 'Fasting, '         
        
        
        
        
        
        
        
        
        
        
        
    row['Circadian Rythm Cohort'] = perc_cohort
    return row    
              
              
```

```python
# calculate all risk factors 
df_CIRC=df_CIRC.apply(interpret_stop_breathing_risk, axis='columns')
df_CIRC=df_CIRC.apply(interpret_trouble_falling_asleep, axis='columns')
df_CIRC=df_CIRC.apply(interpret_low_energy, axis='columns')
df_CIRC=df_CIRC.apply(interpret_sleeping_during_day, axis='columns')
df_CIRC=df_CIRC.apply(interpret_sleep_driving, axis='columns')
df_CIRC=df_CIRC.apply(interpret_snore_loudly, axis='columns')
df_CIRC=df_CIRC.apply(interpret_fasting, axis='columns')
 
df_CIRC=df_CIRC.apply(sum_CIRC_risk_Factors, axis='columns')
df_CIRC=df_CIRC.apply(sum_CIRC_risk_Strat, axis='columns')
df_CIRC=df_CIRC.apply(sum_CIRC_risk_Score, axis='columns')
df_CIRC=df_CIRC.apply(sum_CIRC_risk_Score_wt, axis='columns')
df_CIRC=df_CIRC.apply(sum_CIRC_risk_Cohort, axis='columns')
```

```python
def calc_AIHS_BMI_risk(row):
    # The row is a single Series object which is a single row indexed by column values
    # Let's extract the firstname and create a new entry in the series
    if row['Simulated BMI'] < 18.5: 
        row['AIHS BMI Risk']= 0
    elif row['Simulated BMI'] < 25: 
        row['AIHS BMI Risk']=  0  
    elif row['Simulated BMI']  < 30: 
        row['AIHS BMI Risk']=  1 +  ((row['Simulated BMI']-26) * .1)
    elif row['Simulated BMI']  < 40: 
        row['AIHS BMI Risk']=  2 +  ((row['Simulated BMI']-31) * .1) 
    elif row['Simulated BMI']  < 60:    # Age over 90 non actionable ????
        row['AIHS BMI Risk']=  3 +  ((row['Simulated BMI']-41) * .1)      
    return row
```
