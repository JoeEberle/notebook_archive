```python
def add_slide_with_bullet_points(pptx_file_path, title, bullet_points):
    prs = pptx.Presentation(pptx_file_path)
    slide_layout = prs.slide_layouts[1] # Select the slide layout you want to use

    # Add a slide with the selected layout
    slide = prs.slides.add_slide(slide_layout)

    # Add a title to the slide
    title_shape = slide.shapes.title
    title_shape.text = title

    # Add bullet points to the slide
    body_shape = slide.shapes.placeholders[1]
    tf = body_shape.text_frame
    for bullet_point in bullet_points:
        p = tf.add_paragraph()
        p.text = bullet_point
        p.level = 0

    prs.save(pptx_file_path)
    
testing = True
if testing:
    add_slide_with_bullet_points(presentation_file_name, "What is our platform?", ["Python is the Number one Data Science Language", "PANDAS provides high performance data transformation", "AUTO_ML greatly simplifies machine learning", "AI breakthroughs are occuring EVERYDAY - We need to embed all AI into our tools"])
```
