```python
import random

class card(object):
    
    def __init__(self, suit, card_value):
        self.suit = suit
        self.card_value = card_value
        
    def out(self):
        return " {} of {} ".format(self.card_value,self.suit)
    
class deck(object): 
    import random        
    
    def __init__(self):
        self.cards = []
        self.build()
        
    def build(self):
        for suit in ["Diamonds","Hearts","Clubs","Spades"]:
            for card_value in range(1,14):
                if card_value == 1:
                    card_value = 'Ace'
                if card_value == 13:
                    card_value = 'King'    
                if card_value == 12:
                    card_value = 'Queen'     
                if card_value == 11:
                    card_value = 'Jack'  
                self.cards.append(card(suit,card_value))
                
    def out(self):
        ''' return all of the cards in the deck in the current order '''        
        for card in self.cards:
            card.out()
    
    
    def shuffle(self, number_of_shuffles):
        ''' Randomly shuffle all 52 cards as many time as asked '''
        for shuffle_count in range(0 , number_of_shuffles):     # Shuffle the deck as many times as requested 
            for iterate_card in range(len(self.cards) - 1, 0, -1):    # start at back of deck a shuffle every card 
                r = random.randint( 0 , iterate_card)    # only select a random spot that hasnt been swapped already
                self.cards[r],  self.cards[iterate_card] =  self.cards[iterate_card],  self.cards[r]     #swap positions
                
    def draw(self):
        return self.cards.pop()        

                
# class player(object): 
#     def __init__(self,)    

c1 = card("Diamonds",8)
c2 = card("Clubs",3)
c3 = card("Spades","king")
c4 = card("Hearts","Ace")

# print(c1.out() + ", " + c2.out()  + ", " + c3.out() + "\n" )

card_deck = deck()

# for card in card_deck.cards:
#     print(card.out())
  
num_of_shuffles = 8 
print('shuffling deck')
card_deck.shuffle(num_of_shuffles)


for card in card_deck.cards:   # output the entuire deck of cards 
    print(card.out())
    
print('Drawing top 5 cards')    
    
for counter in range(1, 6):   # draw five top cards 
    draw_card = card_deck.draw()
    print(draw_card.out())    
    
    
```
