```python
from tkinter import * 

root = Tk()
root.title('Simple Calculator')

e = Entry(root, width=35, borderwidth=5)
e.grid(row=0, column=0, columnspan=3, padx=10, pady=10)
e.insert(0, '' )

global first_number  
global second_number 
global calculator_command  
first_number = 0
second_number = 0
calculator_command = '?'


def button_click(key_number):
    current = e.get()  
    e.delete(0, END)
    e.insert(0, current + str(key_number))
    return 

def button_clear():
    e.delete(0, END)
    return 

def button_add():
    first_number = int(e.get()) 
    e.delete(0, END)
    return 

def button_add():
    global first_number  
    global calculator_command          
    first_number = int(e.get()) 
    calculator_command = '+'
    e.delete(0, END)
    e2.delete(0, END)
    e2.insert(0, f'first#{first_number} Command:{calculator_command} ')    
    return 

def button_multiply():
    global first_number  
    global calculator_command          
    first_number = int(e.get()) 
    calculator_command = '*'
    e.delete(0, END)
    e2.delete(0, END)
    e2.insert(0, f'first{first_number} Command:{calculator_command} ')    
    return 

def button_divide():
    global first_number  
    global calculator_command          
    first_number = int(e.get()) 
    calculator_command = '/'
    e.delete(0, END)
    e2.delete(0, END)
    e2.insert(0, f'first#{first_number} Command:{calculator_command} ')    
    return 


def button_subtract():
    global first_number     
    global calculator_command          
    first_number = int(e.get()) 
    calculator_command = '-'
    e.delete(0, END)
    e2.delete(0, END)
    e2.insert(0, f'first#{first_number} Command:{calculator_command} ') 
    return 

def button_equal():
    global first_number  
    global calculator_command      
    second_number = int(e.get()) 
    if calculator_command == '+':
        result = first_number + second_number   
    if calculator_command == '-':
        result = first_number - second_number 
    if calculator_command == '*':
        result = first_number * second_number   
    if calculator_command == '/':
        result = first_number / second_number    
    e.delete(0, END)         
    e.insert(0, result)          
    e2.delete(0, END)    
    e2.insert(0, f'{first_number} {calculator_command} {second_number} = {result}')  
    e2.insert(0, f'first# + {first_number} Command:{calculator_command} second#{second_number} = {result}')     
    first_number = int(e.get()) 
    return 

button_1 = Button(root, text='1', padx=40, pady=20, command=lambda: button_click(1))
button_2 = Button(root, text='2', padx=40, pady=20, command=lambda: button_click(2))
button_3 = Button(root, text='3', padx=40, pady=20, command=lambda: button_click(3))
button_4 = Button(root, text='4', padx=40, pady=20, command=lambda: button_click(4))
button_5 = Button(root, text='5', padx=40, pady=20, command=lambda: button_click(5))
button_6 = Button(root, text='6', padx=40, pady=20, command=lambda: button_click(6))
button_7 = Button(root, text='7', padx=40, pady=20, command=lambda: button_click(7))
button_8 = Button(root, text='8', padx=40, pady=20, command=lambda: button_click(8))
button_9 = Button(root, text='9', padx=40, pady=20, command=lambda: button_click(9))
button_0 = Button(root, text='0', padx=40, pady=20, command=lambda: button_click(0))

button_multiply = Button(root, text='*', padx=39, pady=20, command=button_multiply)
button_divide = Button(root, text='/', padx=39, pady=20, command=button_divide)
button_subtract = Button(root, text='-', padx=39, pady=20, command=button_subtract)
button_add = Button(root, text='+', padx=39, pady=20, command=button_add)
button_equal = Button(root, text='=', padx=40, pady=20, command=button_equal)
button_clear = Button(root, text='C', padx=40, pady=20, command=button_clear)

button_1.grid(row=3, column=0)
button_2.grid(row=3, column=1)
button_3.grid(row=3, column=2)

button_4.grid(row=2, column=0)
button_5.grid(row=2, column=1)
button_6.grid(row=2, column=2)

button_7.grid(row=1, column=0)
button_8.grid(row=1, column=1)
button_9.grid(row=1, column=2)

button_0.grid(row=4, column=0)
button_clear.grid(row=4, column=1)
 
button_add.grid(row=5, column=0)
button_subtract.grid(row=5, column=1)
button_equal.grid(row=5, column=2)

button_multiply.grid(row=6, column=0)
button_divide.grid(row=6, column=1)


e2 = Entry(root, width=35, borderwidth=5, font='arial 10') 
e2.grid(row=7, column=0, columnspan=3, padx=10, pady=10)
e2.insert(0, 'first#' + str(first_number) + ' Command=' + calculator_command + ' second#' + str(second_number))

root.mainloop()
```
